```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_colwidth = 100
```

```python
movies_df = pd.read_csv('data/extended_info.csv', delimiter=",", encoding='utf-8')
```

```python
movies_df.sample(10)
```

```python
movies_df.describe(include='all')
```

```python
language_counts = movies_df['original_language'].value_counts()

# Create the bar plot
plt.bar(language_counts.index, language_counts)

# Add labels and title
plt.xlabel('Original Languages')
plt.ylabel('Counts')
plt.title('Language Counts')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
max_vote_average = movies_df['vote_average'] == movies_df['vote_average'].max()
movies_df[max_vote_average]
```

```python
max_vote_count = movies_df['vote_count'] == movies_df['vote_count'].max()
movies_df[max_vote_count]
```

```python
# Sort the DataFrame by popularity in descending order
sorted_df = movies_df.sort_values('popularity', ascending=False)

# Select the top 10 movies
top_10_popularity = sorted_df.head(10)

# Create the bar plot
plt.bar(top_10_popularity['original_title'], top_10_popularity['popularity'])

# Add labels and title
plt.xlabel('Movie Titles')
plt.ylabel('Popularity')
plt.title('Top 10 Popularity by Movie')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Sort the DataFrame by vote counts in descending order
sorted_df = movies_df.sort_values('vote_count', ascending=False)

# Select the top 10 movies
top_10_counts = sorted_df.head(10)

# Create the bar plot
plt.bar(top_10_counts['original_title'], top_10_counts['vote_count'])

# Add labels and title
plt.xlabel('Movie Titles')
plt.ylabel('Counts')
plt.title('Top 10 Vote Counts Movies')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Filter the DataFrame by original language is English
language_filter = movies_df['original_language'] == 'en'


# Sort the DataFrame by vote average in descending order
sorted_df = movies_df[language_filter].sort_values('vote_average', ascending=False)

# Select the top 10 movies
top_10_vote_average = sorted_df.head(10)

# Create the bar plot
plt.bar(top_10_vote_average['original_title'], top_10_vote_average['vote_average'])

# Add labels and title
plt.xlabel('Movie Titles')
plt.ylabel('Vote Average')
plt.title('Top 10 Vote Average Movies')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
